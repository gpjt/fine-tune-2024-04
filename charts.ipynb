```python
import pandas as pd

# Data from the Markdown table as a list of lists
data = [
    ['python', 1, 6.24, 19692, 52],
    ['torchrun', 1, 5.81, 19692, 56],
    ['torchrun', 2, 5.48, 9846, 30],
    ['torchrun', 3, 5.39, 6564, 20],
    ['torchrun', 4, 5.38, 4924, 15],
    ['torchrun', 5, 5.33, 3940, 12],
    ['torchrun', 6, 5.32, 3282, 10],
    ['torchrun', 7, 5.30, 2814, 9],
    ['torchrun', 8, 5.29, 2462, 8]
]

# Column names from the Markdown table
columns = ['Launcher', '# processes', 'iterations/sec', 'total iterations', 'time (mins)']

# Creating the DataFrame
df = pd.DataFrame(data, columns=columns)
```

```python
import matplotlib.pyplot as plt

plt.xkcd()
plt.rcParams['font.family'] = "xkcd"
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df['# processes'], df['iterations/sec'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('# Processes'.upper())
plt.ylabel('Iterations/sec'.upper())
plt.title('Processes vs Iterations/sec'.upper())
plt.savefig("/tmp/processes-vs-iterations-per-sec.png")
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df['# processes'], df['time (mins)'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('# Processes'.upper())
plt.ylabel('Time (mins)'.upper())
plt.title('Processes vs Time'.upper())
plt.ylim(bottom=0)
plt.savefig("/tmp/processes-vs-time.png")
plt.show()
```
